```python
missinfodf = pd.read_csv("/content/drive/MyDrive/CDC/Twitter/CDCTwitterComplete/CDCTwitterComplete/all_tweets_with_misinfo.csv")
missinfodf
```

```python
df = pd.read_csv('/content/drive/MyDrive/Research-Dr.Ammari/tweetsconcat.csv')
df = df.drop(['index'], axis=1)
df
```

```python
df2 = df["author"].astype('str')
df2 = df2.apply(lambda x: ast.literal_eval(x))
df2 = df2.apply(pd.Series)
df['username'] = pd.Series(df2['username'])
df
```

```python
missinfo = pd.merge(missinfodf,df[['username','id']], left_on='ID', right_on = 'id', how='left')
missinfo = missinfo.drop(['id'], axis=1)
missinfo
```

```python
missinfo = missinfo.drop_duplicates(['Unnamed: 0'])
missinfo = missinfo.reset_index(drop=True)
missinfo
```

```python
missinfo['MBFC cred'].unique()
```

```python
lonly2 = missinfo[missinfo['MBFC cred'] == 'L']
len(lonly2)
```

```python
lonly2['username'].value_counts() #Clearly, people who share low quality tweets share more than just the one
```

```python
lonly2[lonly2['username'] == 'ceciprd']
```

```python
lowqualityusers = list(lonly2['username'].unique())
lowqualityusers = ["@" + x for x in lowqualityusers]
lowqualityusers
```

```python
nodes = pd.read_csv("/content/drive/MyDrive/CDC/ERGM/quadrants with sentiment/nodes_quadrants_sentiment.csv")
nodes = nodes.drop(['Unnamed: 0'], axis=1)
nodes
```

```python
nodes['lowquality'] = 0
nodes
```

```python
for index, row in nodes.iterrows():
  author = str(row['username'])
  if author in lowqualityusers:
    nodes['lowquality'][index] = 1
nodes
```
